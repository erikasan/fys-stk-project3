```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, cross_val_score, KFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
```

### Exploratory data analysis

Descriptive statistics of the data set. 


```python
# load data
cancer_data = load_breast_cancer(as_frame=True)
# data frame for exploratory data analysis
df = cancer_data.frame

# train and test set for machine learning
X, y = cancer_data.data, cancer_data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# scale the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### MLPClassifier

Hva er problemet med denne? FÃ¸rst og fremst, vi mÃ¥ sÃ¸ke etter parametere. Vi mÃ¥ evaluere modellen pÃ¥ hele (CV), kan hende det fÃ¸rste jeg fikk bare var en prediksjon av en enkel del av datasettet. FÃ¥r en score her, sÃ¥ gjÃ¸re en feature selection og motivere dette. Finner ut hvilke features - mest korrelert med responsen, sÃ¥ en grid search pÃ¥ dette.  

 MÃ… SKRIVE OM CORRELATION MATRIX 

```python
mlp = MLPClassifier(hidden_layer_sizes=(50,), max_iter=300, alpha=1e-4,
                   solver="sgd", verbose=10, random_state=1,
                   learning_rate_init=.1)

# train and test set for machine learning
X, y = df.loc[:, df.columns != "target"], df.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# scale the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# fit the data
mlp.fit(X_train_scaled, y_train)

# predict on test data
predictions = mlp.predict(X_test_scaled)

#outcome_labels = sorted(cancer_data.target.unique())
# Confusion Matrix for MLPClassifier
sns.heatmap(
    confusion_matrix(y_test, predictions),
    annot=True,
    fmt="d",
    xticklabels=["Malignant", "Benign"],
    yticklabels=["Malignant", "Benign"]
)

plt.show()
```

```python
# MLP accuracy score
print(accuracy_score(y_test, predictions) * 100) 

# Classification report
print(classification_report(y_test, predictions))

# MLP training set score
print(mlp.score(X_train_scaled, y_train) * 100) 

# MLP testing set score
print(format(mlp.score(X_test_scaled, y_test) * 100)) 
```

### Grid Search

PÃ¥ modell uten feature selection

```python
mlp = MLPClassifier(hidden_layer_sizes=(50,), max_iter=300, alpha=1e-4,
                   solver="sgd", verbose=10, random_state=1,
                   learning_rate_init=.1)

# train and test set for machine learning
X, y = df.loc[:, df.columns != "target"], df.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# scale the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# fit the data
mlp.fit(X_train_scaled, y_train)

# predict on test data
predictions = mlp.predict(X_test_scaled)

outcome_labels = sorted(cancer_data.target.unique())
# Confusion Matrix for MLPClassifier
sns.heatmap(
    confusion_matrix(y_test, predictions),
    annot=True,
    fmt="d",
    xticklabels=outcome_labels,
    yticklabels=outcome_labels
)

plt.show()
```

```python
mlp_gs = MLPClassifier(max_iter=1000)

parameter_space = {
    'hidden_layer_sizes':  [(50), (50, 20), (10,30,)],
    'activation': ['tanh', 'relu', 'logistic'],
    'solver': ['sgd', 'adam', 'lbfgs'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate': ['constant','adaptive', 'invscaling'],
}

clf = GridSearchCV(mlp_gs, parameter_space, n_jobs=-1, cv=5)
clf.fit(X_train_scaled, y_train) # X is train samples and y is the corresponding labels
```

```python
print('Best parameters found:\n', clf.best_params_)
```

### Test new MLP

skewed dataset. Biased mot benign siden det er det det er flest av i datasettet. Treningen pÃ¥ datasettet blir biased mot dette. Malignant har flest falske negative fordi nettverket kommer til Ã¥ gjette pÃ¥ benign enn malignant siden benign har flest samples. 

```python
mlp = MLPClassifier(hidden_layer_sizes=(50, 20), max_iter=1000, activation='logistic', 
                   alpha=0.01, solver="lbfgs", learning_rate="constant")

# train and test set for machine learning
X, y = df.loc[:, df.columns != "target"], df.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# scale the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# fit the data
mlp.fit(X_train_scaled, y_train)

# predict on test data
predictions = mlp.predict(X_test_scaled)

outcome_labels = sorted(cancer_data.target.unique())
# Confusion Matrix for MLPClassifier
sns.heatmap(
    confusion_matrix(y_test, predictions),
    annot=True,
    fmt="d",
    xticklabels=outcome_labels,
    yticklabels=outcome_labels
)

plt.show()
```

```python
# Initialize a KFold instance
k = 5
kfold = KFold(n_splits = k)

scaler = StandardScaler()

X_std = scaler.fit_transform(X)

fe_score = cross_val_score(
    mlp,
    X_std,
    y,
    cv=kfold,
    scoring='f1'
)

print("MLP: F1 after 10-fold cross-validation: {:.2f}% (+/- {:.2f}%)".format(
    fe_score.mean() * 100,
    fe_score.std() * 2
))
```

```python
# MLP accuracy score
print(accuracy_score(y_test, predictions) * 100) 

# Classification report
print(classification_report(y_test, predictions))

# MLP training set score
print(mlp.score(X_train_scaled, y_train) * 100) 

# MLP testing set score
print(format(mlp.score(X_test_scaled, y_test) * 100)) 
```

### Feature selection

GjÃ¸re feature selection, kan bli hÃ¸yere hvis man fjerner features som ikke har noe Ã¥ si. Parameters that are highly correlated - keep one and drop the other? 

```python
corr = df.corr()
plt.subplots(figsize=(26,21))
sns.heatmap(corr, annot=True, fmt='.2g', cmap="coolwarm")
plt.show()
```

```python
corr_target = abs(corr["target"])

relevant_features = corr_target[corr_target>0.3]

df2 = df[relevant_features.keys()]
corr2 = df2.corr()

plt.subplots(figsize=(26,21))
sns.heatmap(corr2, annot=True, fmt='.2g', cmap="coolwarm")
plt.show()
```

```python
#mlp = MLPClassifier(hidden_layer_sizes=(50,), max_iter=300, alpha=1e-4,
#                   solver="sgd", verbose=10, random_state=1,
#                   learning_rate_init=.1)

mlp = MLPClassifier(max_iter=1000, alpha=0.1, activation='logistic', solver='adam', random_state=42)

# train and test set for machine learning
X, y = df2.loc[:, df2.columns != "target"], df2.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# scale the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# fit the data
fit = mlp.fit(X_train_scaled, y_train)

# predict on test data
predictions = mlp.predict(X_test_scaled)

outcome_labels = sorted(cancer_data.target.unique())

# Confusion Matrix for MLPClassifier
sns.heatmap(
    confusion_matrix(y_test, predictions),
    annot=True,
    fmt="d",
    xticklabels=outcome_labels,
    yticklabels=outcome_labels
)

plt.show()
```

```python
# MLP accuracy score
print(accuracy_score(y_test, predictions) * 100) 

# Classification report
print(classification_report(y_test, predictions))

# MLP training set score
print(f"Train accuracy: {mlp.score(X_train_scaled, y_train) * 100}")
# MLP test set score
print(f"Test accuracy: {accuracy_score(y_test, predictions) * 100}")
```
